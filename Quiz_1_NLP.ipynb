Challenge 1: Tokenization : You have a collection of tickets enclosed below. Apply tokenization using python nltk and generate a vocabulary of words in the ticket.  Share the link to the codebase (github link) in the comments section below. Google Colab can be used as an environment.

```
import pandas as pd
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```
ticket_data = pd.read_csv('/content/drive/My Drive/Data/ticket_Data.csv')
```

```
ticket_data.info()
```

```
ticket_desc = ' '.join(ticket_data['Description']).lower()
```

### Normal word_tokenize() can tokenize but it will have noise like duplicate, punctuation, numbers.

```
word_token = word_tokenize(ticket_desc)
```

```
' '.join(set(word_token))
```

### Production grade code need to remove junks from the initial version

```
from nltk.tokenize import RegexpTokenizer
```

```
# First lets create the set of word from the ticket_desc
ticket_desc_set = set(ticket_desc.split(' '))

# Stage 1: To remove the alphanumeric
num_tokenizer = RegexpTokenizer('\w+[0-9]+')
alphanum_vocab1 = num_tokenizer.tokenize(ticket_desc)

alphanum_vocab1_set = set(alphanum_vocab1)

rem_stage1_ticket_desc = ticket_desc_set.difference(alphanum_vocab1_set)
print(rem_stage1_ticket_desc)
```

### You can remove punctuation with the help of re package also

```
import re
```

```
str_stage1_ticket_desc = ' '.join(rem_stage1_ticket_desc)

input_stage2a_ticket_desc = re.sub(r'[\./()\_,-;\[>]', ' ', str_stage1_ticket_desc)
input_stage2b_ticket_desc = re.sub(r'\s+', ' ', input_stage2a_ticket_desc)

result_volcabulary = input_stage2b_ticket_desc.split(' ')
print(result_volcabulary)
```

### Further improvement based on requirement. 
#### Remove the Alphanumeric in other form is some are missed from stage 1

```
num_tokenizer = RegexpTokenizer('[0-9]+\w+')
alphanum_vocab2 = num_tokenizer.tokenize(ticket_desc)
print(alphanum_vocab2)
```

### Remove the Numbers is some are left out

```
num_tokenizer = RegexpTokenizer('[0-9]+')
num_vocab = num_tokenizer.tokenize(ticket_desc)
print(num_vocab)
```

### If requirement is to keep the hyphenated word, below step identify them.

```
hyp_word_tokenizer = RegexpTokenizer('\w+-\w+')
hyp_word_vocab = hyp_word_tokenizer.tokenize(ticket_desc)
print(hyp_word_vocab)
```

```
word_tokenizer = RegexpTokenizer('[\w]+')
word_vocab = word_tokenizer.tokenize(ticket_desc)
print(set(word_vocab))
```
